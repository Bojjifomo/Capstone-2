### **Business Problem Understanding**

**Context**

Pasar mobil bekas di Arab Saudi sangat besar dan kompetitif, dengan banyak dealer maupun platform online seperti syarah.com yang menampilkan ribuan listing mobil setiap hari.


Namun, terdapat beberapa tantangan utama yang dihadapi. Penjual sering kesulitan menentukan harga mobil bekas yang wajar; jika harga terlalu mahal, mobil menjadi sulit terjual, sementara jika terlalu murah, margin keuntungan dapat hilang.

Di sisi lain, pembeli juga kesulitan menilai apakah harga yang ditawarkan sesuai dengan kondisi mobil.


Untuk itu, platform jual beli membutuhkan sistem yang mampu memberikan estimasi harga mobil bekas secara otomatis, sehingga dapat membantu baik penjual maupun pembeli dalam membuat keputusan dengan lebih cepat, tepat, dan akurat.

**Problem Statement**

Stakeholder utama dalam permasalahan ini adalah dealer mobil, platform online jual beli mobil, serta calon pembeli mobil bekas. Permasalahan yang dihadapi adalah penentuan harga mobil bekas yang masih bersifat subjektif dan tidak konsisten.

Kondisi ini dapat menimbulkan kerugian bagi penjual apabila harga yang ditetapkan terlalu murah, atau justru memperlambat penjualan apabila harga terlalu mahal.

Oleh karena itu, tujuan dari proyek ini adalah membangun sebuah model regresi prediktif yang mampu memperkirakan harga mobil bekas berdasarkan atribut-atribut penting seperti merek, model, tahun produksi, jarak tempuh, kapasitas mesin, dan faktor lainnya.

Prediksi harga yang akurat akan sangat bermanfaat bagi dealer dalam menetapkan harga yang kompetitif, membantu meningkatkan kepuasan pelanggan, serta mempercepat proses jual beli di pasar mobil bekas.

**Goals**

Berdasarkan permasalahan tersebut, platform jual beli mobil bekas di Arab Saudi tentu membutuhkan sebuah tool yang mampu memprediksi harga mobil bekas secara akurat.

Alat ini dapat membantu penjual dalam menetapkan harga yang tepat dan kompetitif untuk setiap mobil yang mereka listing-kan, dengan mempertimbangkan berbagai fitur penting seperti merek, model, tahun produksi, kapasitas mesin, jarak tempuh, jenis transmisi, dan opsi tambahan yang dimiliki mobil tersebut.

Bagi penjual, model prediksi ini dapat meminimalkan risiko kerugian akibat salah menentukan harga, sekaligus mempercepat proses penjualan.

Bagi pembeli, adanya estimasi harga yang adil dan transparan akan meningkatkan rasa percaya terhadap platform.

Sedangkan bagi platform jual beli mobil, keberadaan prediction tool ini berpotensi meningkatkan jumlah pengguna aktif (baik penjual maupun pembeli), memperbesar volume transaksi, serta pada akhirnya dapat mendongkrak revenue perusahaan dari fee atau layanan tambahan yang ditawarkan.

**Analytic Approach**

Langkah pertama yang perlu dilakukan adalah menganalisis data mobil bekas untuk menemukan pola dari berbagai fitur yang ada, seperti merek, model, tahun produksi, kapasitas mesin, jarak tempuh, tipe transmisi, serta opsi tambahan yang membedakan satu mobil dengan mobil lainnya.

Analisis ini akan membantu kita memahami faktor-faktor utama yang memengaruhi harga jual mobil bekas di pasar Arab Saudi.

Selanjutnya, kita akan membangun sebuah model regresi yang mampu memprediksi harga mobil bekas berdasarkan fitur-fitur tersebut.

Model ini nantinya dapat berfungsi sebagai prediction tool yang membantu dealer maupun platform jual beli mobil untuk memberikan estimasi harga yang lebih akurat dan objektif.

Dengan adanya model prediksi ini, penjual dapat menetapkan harga yang lebih kompetitif, pembeli bisa mendapatkan gambaran apakah harga mobil wajar atau tidak, dan platform dapat meningkatkan kepercayaan serta pengalaman pengguna secara keseluruhan.


**Metric Evaluation**

Evaluasi metrik yang akan digunakan dalam model prediksi harga mobil bekas adalah RMSE, MAE, dan MAPE.

RMSE (Root Mean Squared Error) digunakan untuk mengukur rata-rata besar kesalahan prediksi dengan memberikan penalti lebih besar pada error yang ekstrem.

MAE (Mean Absolute Error) mengukur rata-rata selisih absolut antara nilai prediksi dengan nilai aktual, sehingga lebih mudah diinterpretasikan dalam konteks bisnis, misalnya rata-rata prediksi harga meleset sebesar X Riyal.

Sementara itu, MAPE (Mean Absolute Percentage Error) memberikan gambaran error dalam bentuk persentase, sehingga dapat menunjukkan rata-rata persentase kesalahan model terhadap harga aktual mobil.

Semakin kecil nilai RMSE, MAE, dan MAPE, maka semakin akurat model dalam memprediksi harga mobil bekas berdasarkan fitur-fitur yang digunakan. Selain itu, metrik R-squared (R²) juga akan dipertimbangkan, terutama jika model linear digunakan sebagai final model. R-squared menunjukkan seberapa baik varians data harga mobil dapat dijelaskan oleh variabel-variabel prediktor.

Nilai R² yang mendekati 1 menunjukkan bahwa model mampu merepresentasikan variasi data dengan baik, meskipun metrik ini kurang valid jika digunakan pada model non-linear.

```python
# Membuat fungsi untuk mengecek outliers dengan metode IQR
def find_anomalies(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit_bawah = q1 - 1.5 * iqr
    limit_atas = q3 + 1.5 * iqr
    print(f'IQR: {iqr}')
    print(f'Limit bawah: {limit_bawah}')
    print(f'Limit atas: {limit_atas}')
    return limit_bawah, limit_atas
```

```python
# Outliers pada fitur Year
print("\nOutliers fitur Year:")
find_anomalies(df_model['Year'])
```

**Choose a Benchmark Model**

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold, cross_val_score
from sklearn.compose import TransformedTargetRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
import numpy as np

#  Definisikan algoritma
lr  = LinearRegression()
knn = KNeighborsRegressor()
dt  = DecisionTreeRegressor(random_state=42)
rf  = RandomForestRegressor(random_state=42, n_estimators=400, n_jobs=-1)
xgb = XGBRegressor(random_state=42, n_estimators=600, max_depth=6, learning_rate=0.08,
                   subsample=0.9, colsample_bytree=0.9, n_jobs=-1)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
models = [
    TransformedTargetRegressor(lr,  func=np.log, inverse_func=np.exp),
    TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp),
    TransformedTargetRegressor(dt,  func=np.log, inverse_func=np.exp),
    TransformedTargetRegressor(rf,  func=np.log, inverse_func=np.exp),
    TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp),
]

cv = KFold(n_splits=5, shuffle=True, random_state=42)


#  Scaler
scaler = StandardScaler(with_mean=False)

# K-Fold
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Mencari algoritma terbaik berdasarkan tiap metrics
mean_rmse, std_rmse = [], []
mean_mae,  std_mae  = [], []
mean_mape, std_mape = [], []

for m in models:
    est = Pipeline([
        ('prep', preprocess),
        ('scaler', scaler),
        ('model', m)
    ])

    # RMSE
    rmse_scores = cross_val_score(est, X_train, y_train, cv=cv,
                                  scoring='neg_root_mean_squared_error')
    mean_rmse.append(rmse_scores.mean()); std_rmse.append(rmse_scores.std())

    # MAE
    mae_scores = cross_val_score(est, X_train, y_train, cv=cv,
                                 scoring='neg_mean_absolute_error')
    mean_mae.append(mae_scores.mean()); std_mae.append(mae_scores.std())

    # MAPE
    mape_scores = cross_val_score(est, X_train, y_train, cv=cv,
                                  scoring='neg_mean_absolute_percentage_error')
    mean_mape.append(mape_scores.mean()); std_mape.append(mape_scores.std())

# Ringkas hasil
model_names = ['Linear (log)', 'KNN (log)', 'DecisionTree (log)', 'RandomForest (log)', 'XGBoost (log)']
results = pd.DataFrame({
    'Model': model_names,
    'CV_RMSE_mean':  [-s for s in mean_rmse],
    'CV_RMSE_std':   [ s for s in std_rmse],
    'CV_MAE_mean':   [-s for s in mean_mae],
    'CV_MAE_std':    [ s for s in std_mae],
    'CV_MAPE_mean':  [-s*100 for s in mean_mape],
    'CV_MAPE_std':   [ s*100 for s in std_mape]
}).sort_values('CV_RMSE_mean')

results
```

```python
# Evaluasi hasil dari 5 kandidat algoritma (Saudi Arabia Used Cars)
results = pd.DataFrame({
    'Model': [
        'Linear Regression (log)',
        'KNN Regressor (log)',
        'DecisionTree Regressor (log)',
        'RandomForest Regressor (log)',
        'XGBoost Regressor (log)'
    ],
    'CV_RMSE_mean':  [-s for s in mean_rmse],
    'CV_RMSE_std':    [s for s in std_rmse],
    'CV_MAE_mean':   [-s for s in mean_mae],
    'CV_MAE_std':     [s for s in std_mae],
    'CV_MAPE_mean':  [-s*100 for s in mean_mape],
    'CV_MAPE_std':   [s*100 for s in std_mape]
}).sort_values('CV_RMSE_mean')

results
```

```python
# Benchmark 2 model terbaik
models = {
    'XGBoost (log)': TransformedTargetRegressor(
        XGBRegressor(
            random_state=42, n_estimators=600, max_depth=6, learning_rate=0.08,
            subsample=0.9, colsample_bytree=0.9, n_jobs=-1
        ),
        func=np.log, inverse_func=np.exp
    ),
    'RandomForest (log)': TransformedTargetRegressor(
        RandomForestRegressor(
            random_state=42, n_estimators=400, n_jobs=-1
        ),
        func=np.log, inverse_func=np.exp
    )
}

score_rmse, score_mae, score_mape = [], [], []

# Prediksi pada test set
for name, mdl in models.items():
    pipe = Pipeline([
        ('prep', preprocess),
        ('scaler', scaler),
        ('model', mdl)
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)

    score_rmse.append(np.sqrt(mean_squared_error(y_test, y_pred)))
    score_mae.append(mean_absolute_error(y_test, y_pred))
    score_mape.append(mean_absolute_percentage_error(y_test, y_pred))

score_before_tuning = pd.DataFrame(
    {'RMSE': score_rmse, 'MAE': score_mae, 'MAPE': score_mape},
    index=models.keys()
)
score_before_tuning
```

**Hyperparameter Tuning**

Karena model XGBoost memiliki performa yang sedikit lebih baik dari RandomForest, maka selanjutnya kita akan menggunakan model XGBoost ini sebagai model akhir. Pada tahap ini, kita akan melakukan hyperparameter tuning pada model XGBoost dengan harapan dapat meningkatkan performa model.

```python
# space hyperparameter XGBoost (ringan & realistis)
hyperparam_space_xgb = {
    'model__regressor__max_depth':        list(range(3, 11)),
    'model__regressor__learning_rate':    np.linspace(0.02, 0.2, 10),
    'model__regressor__n_estimators':     list(range(150, 451, 50)),
    'model__regressor__subsample':        np.linspace(0.6, 1.0, 5),
    'model__regressor__colsample_bytree': np.linspace(0.6, 1.0, 5),
    'model__regressor__gamma':            list(range(0, 6)),
    'model__regressor__reg_alpha':        np.logspace(-3, 1, 8),
    'model__regressor__reg_lambda':       np.logspace(-3, 1, 8),
}

# estimator baseline
xgb_base = TransformedTargetRegressor(
    XGBRegressor(random_state=42, tree_method='hist', n_jobs=-1),
    func=np.log, inverse_func=np.exp
)

xgb_pipe = Pipeline([
    ('prep', preprocess),
    ('scaler', StandardScaler(with_mean=False)),  # aman untuk OHE sparse
    ('model', xgb_base)
])

# RandomizedSearch (lebih cepat dari Grid)
cv = KFold(n_splits=3, shuffle=True, random_state=42)
xgb_search = RandomizedSearchCV(
    estimator=xgb_pipe,
    param_distributions=hyperparam_space_xgb,
    n_iter=30,
    scoring='neg_root_mean_squared_error',
    cv=cv,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

xgb_search.fit(X_train, y_train)
print("Best RMSE (CV):", -xgb_search.best_score_)
print("Best params:", xgb_search.best_params_)
```

Hasil tuning menunjukkan bahwa XGBoost (log) memberikan performa terbaik dengan RMSE CV ≈ 30,614. Kombinasi terbaik diperoleh saat model menggunakan subsample 80% dan reg_lambda sangat kecil (0.001). Artinya, model jadi lebih akurat setelah disesuaikan dengan parameter ini.


```python
# estimator XGBoost dibungkus log-transform
xgb_ttr = TransformedTargetRegressor(
    regressor=XGBRegressor(random_state=42, tree_method='hist', n_jobs=-1, verbosity=0),
    func=np.log, inverse_func=np.exp
)

# pipeline: preprocess -> scaler (aman utk OHE sparse) -> model ===
scaler = StandardScaler(with_mean=False)
estimator_xgb = Pipeline([
    ('prep', preprocess),        # ColumnTransformer yang sudah dibuat
    ('scaler', scaler),
    ('model', xgb_ttr)
])

# CV
crossval = KFold(n_splits=5, shuffle=True, random_state=42)

# RandomizedSearch
random_xgb = RandomizedSearchCV(
    estimator=estimator_xgb,
    param_distributions=hyperparam_space_xgb,   # gunakan dict dengan key 'model__regressor__...'
    n_iter=50,
    cv=crossval,
    scoring={
        'rmse': 'neg_root_mean_squared_error',
        'mae':  'neg_mean_absolute_error',
        'mape': 'neg_mean_absolute_percentage_error'
    },
    refit='rmse',        # optimasi berdasarkan RMSE
    n_jobs=-1,
    verbose=1,
    random_state=42
)
```

```python
# Fitting data training untuk mencari parameter terbaik
random_xgb.fit(X_train, y_train)
```

```python
# Ringkas hasil RandomizedSearchCV ke dalam DataFrame yang enak dibaca
cv = pd.DataFrame(random_xgb.cv_results_)

# ambil semua kolom parameter (prefiks 'param_...') agar tidak hardcode
param_cols = [c for c in cv.columns if c.startswith('param_')]

# pilih kolom metrik & rank utama
metric_cols = [
    'rank_test_rmse', 'mean_test_rmse', 'std_test_rmse',
    'rank_test_mae',  'mean_test_mae',  'std_test_mae',
    'rank_test_mape', 'mean_test_mape', 'std_test_mape'
]

# beberapa versi sklearn menulis nama kolom sesuai key scoring.
# Pastikan kamu memakai scoring={'rmse':..., 'mae':..., 'mape':...} seperti sebelumnya.
# Jika tidak, sesuaikan 'rmse/mae/mape' menjadi nama yang kamu pakai.

# sort by rank (RMSE → MAE → MAPE)
out = cv.sort_values(
    by=['rank_test_rmse', 'rank_test_mae', 'rank_test_mape']
)[param_cols + metric_cols].head(10).copy()

# ubah nilai neg_* menjadi positif (lebih intuitif)
for col in ['mean_test_rmse','std_test_rmse','mean_test_mae','std_test_mae','mean_test_mape','std_test_mape']:
    out[col] = -out[col]

# opsional: tampilkan MAPE dalam persen
out['mean_test_mape'] = out['mean_test_mape'] * 100
out['std_test_mape']  = out['std_test_mape']  * 100

# rename kolom agar lebih jelas
out = out.rename(columns={
    'rank_test_rmse':'rank_RMSE', 'mean_test_rmse':'RMSE_mean', 'std_test_rmse':'RMSE_std',
    'rank_test_mae':'rank_MAE',   'mean_test_mae':'MAE_mean',   'std_test_mae':'MAE_std',
    'rank_test_mape':'rank_MAPE', 'mean_test_mape':'MAPE_mean(%)', 'std_test_mape':'MAPE_std(%)'
})

out.head(10)
```
